 Described Columns

---------------------------------------------------------------------------------------------------------------
--> item_name: This column contains the name of the lost item.

-->  description: Provides additional details or a description of the lost item.

--> receiving_date: Represents the date when the lost item was received at the station.

# Cleaning data

```python
# Remove rows with missing values
df.dropna(inplace=True)

# Fill missing values with a specific value
df.fillna(0, inplace=True)
```

```python
# Convert date strings to datetime objects
df['receiving_date'] = pd.to_datetime(df['receiving_date'], format='%d/%m/%Y')

# Convert time strings to datetime objects
df['receiving_time'] = pd.to_datetime(df['receiving_time']).dt.time
```

```python
# Standardize station names
df['station_name'] = df['station_name'].str.upper()
```

# Bar Chart

A bar plot shows comparisons among discrete categories.

A bar chart is used to display a collection of categories on one axis and another axis represents measured value.

Bar charts can be plotted vertically or horizontally.

```python
top_stations = df["station_name"].value_counts().nlargest(40)
top_stations.plot(kind='bar', figsize=(10,5))
plt.xlabel("Station Name")
plt.ylabel("Frequency")
plt.title("Top 40 Most Frequent Stations")
plt.grid(True)  # Add grid lines
plt.show()
```

# Histogram

A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval. 

```python
top_stations = df["station_name"].value_counts().nlargest(40)
top_stations.plot(kind='hist', figsize=(10,5))
plt.xlabel("Station Name")
plt.ylabel("Frequency")
plt.title("Top 40 Most Frequent Stations")
plt.grid(True)  # Add grid lines
plt.show()
```

# Scatter Plot

```python
# For example, let's say i want to create a scatter plot of item_quantity vs receiving_date

plt.figure(figsize=(10, 6))
plt.scatter(df['receiving_date'], df['item_quantity'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Item Quantity vs Receiving Date')
plt.xlabel('Receiving Date')
plt.ylabel('Item Quantity')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Line Chart

```python
plt.figure(figsize=(10, 6))#This code will create a line chart where the x-axis represents the receiving date and the y-axis represents the item quantity. 
plt.plot(df['receiving_date'], df['item_quantity'], marker='o', linestyle='-')
plt.title('Line Chart of Item Quantity Over Time')
plt.xlabel('Receiving Date')
plt.ylabel('Item Quantity')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```
