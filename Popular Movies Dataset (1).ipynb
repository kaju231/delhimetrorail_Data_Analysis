# Title: Popular Movies Dataset

Description:
This dataset contains information about popular movies obtained from The Movie Database (TMDb) API and converted into a CSV file. The data includes details such as movie title, overview, original language, release date, popularity, vote count, and vote average. The dataset is suitable for analysis, visualization, and machine learning projects related to movie trends, popularity, and user ratings.

Columns:

title: Title of the movie.
overview: Brief overview or summary of the movie.
original_language: Original language of the movie.
release_date: Release date of the movie.
popularity: Popularity score of the movie on TMDb.
vote_count: Number of votes received by the movie.
vote_average: Average rating given to the movie.
Data Source: The data was obtained from The Movie Database (TMDb) API and converted into a CSV file.

```python

import requests
import pandas as pd

# Create an empty DataFrame
final = pd.DataFrame()

# Loop through the pages
for j in range(1, 501):
    # Make the API request
    response = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=390baf909d9302d111c0ed0f1c338527&language=en-US&page={}".format(j)).json()
    
    # Initialize lists to store data for each page
    popularity = []
    vote_count = []
    original_language = []
    title = []
    vote_average = []
    overview = []
    release_date = []

    # Loop through the results and extract data
    for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        title.append(i['title'])
        vote_average.append(i['vote_average'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])

    # Create a dictionary with the lists
    d = {'title': title,
         'overview': overview,
         'original_language': original_language,
         'release_date': release_date,
         'popularity': popularity,
         'vote_count': vote_count,
         'vote_average': vote_average}

    # Create DataFrame for the current page
    df = pd.DataFrame(d)

    # Append the DataFrame to the final DataFrame
    final = final.append(df, ignore_index=True)

# Display the final DataFrame
print(final)  # fetch data
```

# Bar Chart

```python
# Visualize the top movies based on vote count
plt.figure(figsize=(10, 6))
plt.barh(top_vote_count['title'], top_vote_count['vote_count'], color='skyblue')
plt.xlabel('Vote Count')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies by Vote Count')
plt.gca().invert_yaxis()  # Invert y-axis to show the movie with the highest vote count at the top
plt.show()
```

```python
# Count the number of movies for each original language
language_counts = df['original_language'].value_counts()
language_counts
```

# Pie Chart

```python
# Define custom colors for the pie chart slices
colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue', 'lightpink', 'lightyellow', 'lightgrey']

# Create a pie chart with custom colors
plt.figure(figsize=(8, 8))
plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Distribution of Movies by Original Language')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[['popularity', 'vote_count', 'vote_average']].corr()

print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Sort the DataFrame by 'popularity' for better visualization
df.sort_values(by='popularity', inplace=True)
#I first sort the DataFrame by 'popularity' to ensure that the points are plotted in ascending order of popularity.
#we create a scatter plot using scatter() to plot the data points.
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['popularity'], df['vote_count'], color='skyblue', label='Data Points')
plt.plot(df['popularity'], df['vote_count'], color='red', linewidth=2, label='Line of Best Fit')
plt.title('Scatter Plot with Line of Best Fit')
plt.xlabel('Popularity')
plt.ylabel('Vote Count')
plt.legend()
plt.grid(True)
plt.show()
```
